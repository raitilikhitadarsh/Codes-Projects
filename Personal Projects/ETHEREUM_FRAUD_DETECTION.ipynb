# HANDLING IMBALANCE

SMOTE stands for Synthetic Minority Oversampling Technique. This is a statistical technique for increasing the number of cases in your dataset in a balanced way. The module works by generating new instances from existing minority cases that you supply as input.

SMOTE - https://www.geeksforgeeks.org/ml-handling-imbalanced-data-with-smote-and-near-miss-algorithm-in-python/

```python
# import SMOTE module from imblearn library
from imblearn.over_sampling import SMOTE

# Instantiate
sm = SMOTE(random_state = 2)

#fit on train data
X_train_res, y_train_res = sm.fit_sample(X_train, y_train.ravel())

# printthe shape of X_train After OverSampling
print('After OverSampling, the shape of train_X: {}'.format(X_train_res.shape))


# printthe shape of y_train After OverSampling
print('After OverSampling, the shape of train_y: {} \n'.format(y_train_res.shape))


print("After OverSampling, counts of label '1': {}".format(sum(y_train_res == 1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train_res == 0)))
```

```python
def generate_feature_scores_df(X,Score):
    feature_score=pd.DataFrame()
    for i in range(X.shape[1]):
        new =pd.DataFrame({"Features":X.columns[i],"Score":Score[i]},index=[i])
        feature_score=pd.concat([feature_score,new])
    return feature_score
```

```python
# Visualize importance of all the features in the dataset for the prediction
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

rfe = RFE(estimator = RandomForestClassifier(), n_features_to_select=18)

fit = rfe.fit(X,y)
    
    # creating dataframe for feature name and feature importance
  


cols = X.columns[fit.get_support(indices=True)]
# Reduce X to the selected features of anova model using tranform 

X_new = fit.transform(X)
X_new = pd.DataFrame(X_new)
X_new.columns = cols


X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X_new,y,test_size = 0.2)




X_train_res, y_train_res = sm.fit_sample(X_train_new, y_train_new.ravel())
pipeline.fit(X_train_new, y_train_new)

# printthe shape of X_train After OverSampling
print('After OverSampling, the shape of train_X: {}'.format(X_train_res.shape))


# printthe shape of y_train After OverSampling
print('After OverSampling, the shape of train_y: {} \n'.format(y_train_res.shape))


print("After OverSampling, counts of label '1': {}".format(sum(y_train_res == 1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train_res == 0)))

feature_scores_df = generate_feature_scores_df(X_new,fit.ranking_)
# grouping all data and sorting in descending order
sorted_feature_importance = feature_scores_df.sort_values(by = 'Score', ascending = False)
sorted_feature_importance

    # ploting feature importance data using boxenplot
fig = plt.subplots(figsize = (10,10)) 
sns.boxenplot(x = 'Score', y = 'Features', data = sorted_feature_importance)
plt.show()
```

# Hyper parameter tuning

### RandomizedSearchCV

RandomizedSearchCV is very useful when we have many parameters to try and the training time is very long.

1. The first step is to write the parameters that we want to consider
2. From these parameters select the best ones.(which are printed in output)

```python
# Helper function to perform hyper parameter tunning with RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV

def helper(model, parameters):
    randm_src = RandomizedSearchCV(estimator=model, param_distributions = parameters, cv = 3, n_iter = 50)
    randm_src.fit(X_train, y_train)
    return randm_src.best_params_
  # Random search of parameters, using 3 fold cross validation, 
  # search across 100 different combinations, and use all available cores

  # Fit the random search model
```

```python
%%time
# for random forest classifier
from scipy.stats import uniform as sp_randFloat
from scipy.stats import randint as sp_randInt
# create parameters dict for tunning
params = {
    'max_leaf_nodes': sp_randInt(4, 10),
    'n_estimators' : sp_randInt(100, 1000),
    'max_depth'    : sp_randInt(4, 10)
}

best_params = helper(rfc,params)

# passing data for hyper parameter tunning with Randomized search cv
best_params
```
