### Data Analysis regarding COVID-19 pandemic from 31st Dec 2019 to 19th Oct 2020 for India and US.

Source- https://www.kaggle.com/shashwatwork/impact-of-covid19-pandemic-on-the-global-economy

Dataset Format- CSV File

Comments- This dataset contains the data of total cases, total deaths, stringency index, population, country and iso_code which gives us a glimpse of how the world got affected from the ongoing Covid-19 pandemic and how their respective governments reacted within the Timeframe 31-12-2019 to 19-10-2020

```python
#importing libraries, opening and reading the csv file separated by comma(,)
import csv

f = open("./raw_data.csv")

all_lines = csv.reader(f, delimiter = ',')

#to know the attributes that we're working with
next(all_lines)
```

```python
#to convert the entities into dictionary objects which will be useful for us to access "humanely"
dataset = []
header = ['Iso Code', 'Location', 'Date', 'Total Cases', 'Total Deaths', 'Stringency Index', 'Population', 
          'GDP', 'HD Index']
for line in all_lines:
    d = dict(zip(header, line))
    dataset.append(d)
```

```python
#checking if the above cell worked with success or not
len(dataset)
```

```python
#checking whether the provided keys are shown alongside the data or not
dataset[0]
```

```python
#Filtering total reported cases and corresponding dates for the country India
Cases_India = [d['Total Cases'] for d in dataset if d['Location'] == "India"]
Date_Ind = [d['Date'] for d in dataset if d['Location'] == "India"]
```

## To visualise total cases reported for the country India

```python
#rcParams helps us to zoom in and out of our visualisation
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['figure.figsize'] = 10,10
plt.xlabel("Time")
plt.ylabel("Total cases")
plt.title("Covid-19 cases in India")
plt.plot(Date_Ind, Cases_India)
plt.xticks([])
plt.yticks([])
```

```python
#Filtering total reported cases and corresponding dates for the country United States
Cases_US = [d['Total Cases'] for d in dataset if d['Location'] == "United States"]
Dates_US = [d['Date'] for d in dataset if d['Location'] == "United States"]
```

## To juxtapose the difference of cases reported in countries US and India (US exceeds India with a vast difference)

```python
plt.plot(Date_Ind, Cases_India, 'r')
plt.plot(Date_Ind, Cases_US, 'g');
plt.xlabel("Time")
plt.ylabel("Total Cases")
plt.title("Total cases reported in the countries USA (green) and India (red)")
plt.yticks([])
plt.xticks([])
```

## To measure the strictness of the govt with increasing number of cases (Stringency Index increases with increase in number of cases for India)

```python
SI_Ind = [d['Stringency Index'] for d in dataset if d['Location'] == "India"]
plt.plot(Date_Ind, SI_Ind)
plt.plot(Date_Ind, Cases_India);
plt.title("Stringency Index of Govt. of India (blue) with increase in reportings of cases (orange)")
plt.yticks([])
plt.xticks([])
plt.xlabel("Time")
```

## To compare the stringency indices of India and US (since US has more number reported cases, the govt. of US is stricter than that of India)

```python
SI_US = [d['Stringency Index'] for d in dataset if d['Location'] == "United States"]
plt.plot(Date_Ind, SI_Ind, color = 'r')
plt.plot(Date_Ind, SI_US, color = 'g');
plt.xlabel("Time")
plt.ylabel("Stringency Indices of Ind and US")
plt.title("Strictness of Govt. of US (green) and India (red)")
plt.xticks([])
```
